```python
from scipy.stats import spearmanranr
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.random.uniform(0, 3, 500)
y = np.random.uniform(0, 1, 500)
plt.scatter(x, y);
```

```python
np.corrcoef(x, y)[0,1]
```

```python
spearmanr(x, y).correlation
```

- No están correlacionadas
- No pueden ser causales

```python
x = np.random.uniform(0, 3, 300)
y = 6 * x + 1 + np.random.randn(300)
plt.scatter(x, y);
```

```python
np.corrcoef(x, y)[0,1]
```

```python
spearmanr(x, y).correlation
```

- Están correlacionadas fuertemente y positivamente
- No podemos indicar causalidad, podría ser una relación espuria (derivada de terceros factores no representados)

```python
x = np.random.uniform(0, 3, 300)
y = - 3  * x + 1 + 3 * np.random.randn(300)
plt.scatter(x, y);
```

```python
np.corrcoef(x, y)[0,1]
```

```python
spearmanr(x, y).correlation
```

- Correlación negativa débil
- No podemos indicar causalidad, podría ser una relación espuria (derivada de terceros factores no representados)

- **CAUSALIDAD IMPLICA CORRELACIÓN**
- **CORRELACION NO IMPLICA CAUSALIDAD** (tomarlo con pizas)

Como causalidad implica correlación, **si no hay correlación, no hay causalidad**

```python
x = np.random.uniform(0, 3, 300)
y = np.exp(2 * x) + 1 + 2 * np.random.randn(300)
plt.scatter(x, y);
```

```python
np.corrcoef(x, y)[0,1]
```

```python
spearmanr(x, y).correlation
```
